## Statistical methods of Decision making:

To gain some insight in our data in a definite manner we perform statistical test to gain firm evidence.

#### First we will try to find if there is any difference in the price because of area of the house. 

```python
l1 = []
df['total_area15'] = df['living_measure15']+df['lot_measure15']
for i in df['total_area15']:
    if(i<8000):
        l1.append('Small')
    elif(i>=8000 and i<16000):
        l1.append('Medium')
    else:
        l1.append('Large')
df['Area_type'] = l1
```

```python
subset =  df.loc[df['total_area15']<=24000,['Area_type','price']]
rows = subset.sample(frac=0.25)
```

#### Since there are more than 2 groups here we will perform ANOVA test.
#### We wish to determine if there is dependence of price on area of the house.

```python
print('Different groups present are ',rows['Area_type'].unique())
```

## Test for assumptions:
1. Residuals are normally distributed
2. Homogenity of variances
3. Observations are sampled independently from each other

```python
rows
```

### 1. Shapiro wilk test (Normality of Residuals):


We need to create a model before applying shapiro wilks test

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
model = ols('price ~ C(Area_type)',data=rows).fit()
```

H0 : Data is drawn from normal distribution
  
H1 : Data is not drawn from a normal distribution

```python
import scipy.stats as stats
w, pvalue = stats.shapiro(model.resid)
print(w, pvalue)
```

p-value of zero indicates that we reject null hypothesis. 

Hence, data is not drawn from a normal distribution.

## Kruskal Wallis test 

https://www.statisticshowto.com/kruskal-wallis/

```python
s1 = np.array([rows.loc[rows['Area_type']=='Small','price']])
s2 = np.array([rows.loc[rows['Area_type']=='Medium','price']])
s3 = np.array([rows.loc[rows['Area_type']=='Large','price']])
```

```python
s1 = np.ravel(s1)
s1 = s1[:326]
s2 = np.ravel(s2)
s2 = s2[:326]
s3 = np.ravel(s3)
s3 = s3[:326]
print(s1.shape)
print(s2.shape)
print(s3.shape)
```

```python
from scipy.stats import kruskal
stat,p = kruskal(s1,s2,s3)
print('Statistics=%.3f, p=%.3f' % (stat, p))
alpha = 0.05
if p > alpha:
    print('Same distributions (fail to reject H0)')
else:
    print('Different distributions (reject H0)')
```

#### We can see that by performing the kruskal wallis test we found that the groups present in the data for area type has significant difference.

#### We can interpret from the result that small, medium and large house types do not have similar values of price i.e there is a difference between their prices. Hence, we can conclude that for different area types the prices are different

We have performed visualization of the data where we can understand few points:

## Mann-Whitney U test:

We perform this non-parametric test to check two independent samples of price for yr_sold because of failure of normality test

```python
x1 = np.array([row.loc[row['yr_sold']==2015,'price']])
x2 = np.array([row.loc[row['yr_sold']==2014,'price']])
fig, axis = plt.subplots(2)
sns.distplot(x1,ax=axis[0])
sns.distplot(x2,ax=axis[1])
```

```python
x1 = np.ravel(x1)
x1 = x1[:1000]
x2 = np.ravel(x2)
x2 = x2[:1000]
```

H0: No significant difference in price because of yr_sold

H1: Significant difference in price because of yr_sold

```python
from scipy.stats import mannwhitneyu
stat, p = mannwhitneyu(x1, x2)
print('Statistics=%.3f, p=%.3f' % (stat, p))
alpha = 0.05
if p > alpha:
    print('Same distribution (fail to reject H0)')
else:
    print('Different distribution (reject H0)')
```

## Multicollinearity detection and removal:

Structural multicollinearity can be removed by standardizatoin (between living_measure,live_area_change and total_measure and total_area_change)

Here we are going to remove Data Multicollinearity if present

```python
data.head()
```

```python
# Import library for VIF
from statsmodels.stats.outliers_influence import variance_inflation_factor
def calc_vif(X):
    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

    return(vif)
```

```python
X = data.drop(columns=['price'])
calc_vif(X)
```
